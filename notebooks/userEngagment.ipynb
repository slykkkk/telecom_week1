```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the cluster metrics data
cluster_metrics = pd.read_csv("../data/engagement_cluster_stats.csv")

# Define the metrics to visualize
metrics = ['Sessions Frequency', 'Session Duration', 'Total DL Traffic', 'Total UL Traffic']

# Create a figure to hold all subplots
fig, axs = plt.subplots(len(metrics), 1, figsize=(10, 8))

# Iterate over each metric
for i, metric in enumerate(metrics):
    # Extract cluster data for the metric
    cluster_data = cluster_metrics.set_index('Cluster')[metric]
    cluster_data.dropna(inplace=True)  # Drop NaN values
    
    # Plot the data
    axs[i].plot(cluster_data, marker='o', linestyle='-')
    
    # Add labels and title
    axs[i].set_title(f'{metric} by Cluster')
    axs[i].set_xlabel('Cluster')
    axs[i].set_ylabel(metric)
    axs[i].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Application': ['App A', 'App B', 'App C'],
    'Total Traffic (MB)': [100, 80, 60]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.barh(df['Application'], df['Total Traffic (MB)'], color='skyblue')

# Add labels and title
plt.xlabel('Total Traffic (MB)', fontsize=12, fontweight='bold')
plt.ylabel('Application', fontsize=12, fontweight='bold')
plt.title('Top 3 Most Used Applications', fontsize=14, fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```
