```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv('xdr_data.csv')

# Select two or three relevant numerical columns
relevant_columns = ['Dur. (ms)', 'Total UL (Bytes)', 'Total DL (Bytes)']  # Adjust as needed

# Check if df contains the selected numerical columns
if all(col in df.columns for col in relevant_columns):
    # Create a figure and axes
    fig, ax = plt.subplots(len(relevant_columns), 1, figsize=(8, 6 * len(relevant_columns)), sharex=True)

    # Iterate through each selected column and create Q-Q plot
    for i, column in enumerate(relevant_columns):
        stats.probplot(df[column], dist="norm", plot=ax[i])
        ax[i].set_title(f'Q-Q Plot for {column}')
        ax[i].set_ylabel('Observed Quantiles')
        ax[i].set_xlabel('Theoretical Quantiles')
        ax[i].grid(True)

    # Adjust layout
    plt.tight_layout()
    plt.show()
else:
    print("DataFrame does not contain all selected numerical columns.")
```

```python
import numpy as np

# Define relevant variables for PCA
data = df[['Social Media DL (Bytes)', 'Social Media UL (Bytes)',
           'Google DL (Bytes)', 'Google UL (Bytes)',
           'Email DL (Bytes)', 'Email UL (Bytes)',
           'Youtube DL (Bytes)', 'Youtube UL (Bytes)',
           'Netflix DL (Bytes)', 'Netflix UL (Bytes)',
           'Gaming DL (Bytes)', 'Gaming UL (Bytes)',
           'Other DL (Bytes)', 'Other UL (Bytes)']]

# Standardize the data
data_std = (data - data.mean()) / data.std()

# Perform PCA
cov_matrix = np.cov(data_std.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Calculate explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

# Print explained variance ratio
print(explained_variance_ratio)

# Print principal components
print(eigenvectors)

# Scree plot
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load your dataset into a DataFrame
df = pd.read_csv('xdr_data.csv')

# Select relevant columns for dimensionality reduction
relevant_columns = ['Bearer Id', 'Dur. (ms)', 'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 
                    'Other DL (Bytes)', 'Total UL (Bytes)', 'Total DL (Bytes)']

# Separate features and target variable (if applicable)
X = df[relevant_columns]

# Impute missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Instantiate PCA and specify the number of components
pca = PCA(n_components=2)  # You can adjust the number of components as needed

# Fit PCA to the scaled data
pca.fit(X_scaled)

# Transform the data to the new feature space
X_pca = pca.transform(X_scaled)

# Visualize the data in the reduced feature space
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Reduced Dimensionality')
plt.grid(True)
plt.show()
```
