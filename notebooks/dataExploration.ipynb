```python
# Check data type of 'Dur. (ms)'
print(df['Dur. (ms)'].dtype)

# Check for missing values in 'Dur. (ms)'
print(df['Dur. (ms)'].isnull().sum())

# Drop rows with missing values in 'Dur. (ms)'
df.dropna(subset=['Dur. (ms)'], inplace=True)

# Use pandas qcut to segment into deciles based on 'Dur. (ms)'
try:
    df['Decile_Class'] = pd.qcut(df['Dur. (ms)'], q=10, labels=False, duplicates='drop')
except ValueError as e:
    print("Error:", e)

# Compute total data (DL+UL) per decile class
total_data_per_decile = df.groupby('Decile_Class').agg({
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
})

# Print total data per decile class
print(total_data_per_decile)
```

```python
import numpy as np

# Define relevant variables for PCA
data = df[['Social Media DL (Bytes)', 'Social Media UL (Bytes)',
           'Google DL (Bytes)', 'Google UL (Bytes)',
           'Email DL (Bytes)', 'Email UL (Bytes)',
           'Youtube DL (Bytes)', 'Youtube UL (Bytes)',
           'Netflix DL (Bytes)', 'Netflix UL (Bytes)',
           'Gaming DL (Bytes)', 'Gaming UL (Bytes)',
           'Other DL (Bytes)', 'Other UL (Bytes)']]

# Standardize the data
data_std = (data - data.mean()) / data.std()

# Perform PCA
cov_matrix = np.cov(data_std.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Calculate explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

# Print explained variance ratio
print(explained_variance_ratio)

# Print principal components
print(eigenvectors)

# Scree plot
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```
